```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVR
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from lightgbm import LGBMRegressor
import xgboost as xgb
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import RandomizedSearchCV
from sklearn.inspection import permutation_importance
from sklearn.metrics import mean_absolute_error, mean_squared_error
from scipy.stats import ttest_rel, wilcoxon
import warnings
warnings.filterwarnings('ignore')

# Set visual aesthetics for the notebook
sns.set_theme(style="whitegrid", palette="muted")
plt.rcParams.update({
    "font.family": "sans-serif",
    "font.size": 12,
    "figure.dpi": 100,
    "savefig.bbox": "tight"
})
```

## 1. Data Ingestion & Chronological Split
We load the 14-feature fundamental dataset for 105 hydropower companies. We engineer the Target variable as the **Percentage Growth Rate of Free Cash Flow**.

Crucially, to avoid **Lookahead Bias**, we utilize a strict chronological split rather than a random split. The ML models train exclusively on historical data ($<2023$) to predict the future test set ($\ge 2023$).


```python
# Load Master Dataset
df = pd.read_csv('data/master_ml_dataset.csv')
df = df.sort_values(by=["Ticker", "Year_Int"]).reset_index(drop=True)

# Engineer Target Variable: % Growth in Free Cash Flow
df["FCF_NextYear"] = df.groupby("Ticker")["OperatingCashFlow_Proxy"].shift(-1)
df["Target_FCF_Growth"] = (df["FCF_NextYear"] - df["OperatingCashFlow_Proxy"]) / (df["OperatingCashFlow_Proxy"].abs() + 1)
df = df.dropna(subset=["Target_FCF_Growth"]).copy()

# Winsorize the target to handle extreme distress outliers
lower = df["Target_FCF_Growth"].quantile(0.05)
upper = df["Target_FCF_Growth"].quantile(0.95)
df["Target_FCF_Growth"] = df["Target_FCF_Growth"].clip(lower, upper)
df = df.fillna(0)

features = [
    "Revenue", "Debt_Ratio", "ROA", "Working_Capital", "CapEx",
    "Installed_Capacity_MW", "PLF", "License_Term_Years",
    "GDP_Growth", "Inflation", "Interest_Rate", 
    "Years_Since_COD", "Pre_Post_COD", "EBIT_Margin"
]

# Chronological Train-Test Split (Avoid Lookahead Bias)
train_df = df[df["Year_Int"] < 2023]
test_df = df[df["Year_Int"] >= 2023]

X_train = train_df[features]
y_train = train_df["Target_FCF_Growth"]
X_test = test_df[features]
y_test = test_df["Target_FCF_Growth"]

print(f"Training Firm-Years: {len(X_train)}")
print(f"Testing Firm-Years: {len(X_test)}")
```

## 2. Exploratory Data Analysis: Correlation Heatmap
Before modeling, it is critical in corporate finance to examine the multicollinearity between the 14 operational and macroeconomic variables.


```python
plt.figure(figsize=(14, 10))
corr_matrix = train_df[features + ["Target_FCF_Growth"]].corr()
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", center=0, square=True, linewidths=.5, cbar_kws={"shrink": .75})
plt.title("Correlation Matrix of Features and Target Variable", fontsize=16)
plt.show()
```

## 3. The ML Showdown: Training the Algorithms
We will test 6 distinct architectures against a Naive Traditional Baseline (which assumes 0% growth). We will evaluate them using **Reconstructed Mean Absolute Error (MAE)**. We use standard scaling since SVR and KNN are distance-based algorithms.


```python
# 1. Traditional Baseline
trad_preds = test_df["OperatingCashFlow_Proxy"]
actual_fcf = test_df["FCF_NextYear"]
trad_mae = mean_absolute_error(actual_fcf, trad_preds)
baseline_errors = np.abs(actual_fcf - trad_preds)
print(f"Traditional Baseline MAE: Rs {trad_mae:,.0f}")

# 2. Scale Features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

results = {}
base_fcf_test = test_df["OperatingCashFlow_Proxy"]

def evaluate_model(name, model, param_grid):
    search = RandomizedSearchCV(model, param_grid, n_iter=15, scoring="neg_mean_absolute_error", cv=3, n_jobs=-1, random_state=42)
    search.fit(X_train_scaled, y_train)
    best_model = search.best_estimator_
    pred_growth = best_model.predict(X_test_scaled)
    # Reconstruct Absolute Cash Flow
    abs_preds = base_fcf_test + (base_fcf_test.abs() + 1) * pred_growth
    mae = mean_absolute_error(actual_fcf, abs_preds)
    results[name] = mae
    print(f"{name} MAE: Rs {mae:,.0f}")
    return best_model, abs_preds

print("\nTuning SVR...")
svr_params = {"C": [0.1, 1, 10, 100], "gamma": ["scale", "auto", 0.1, 0.01], "epsilon": [0.01, 0.1, 1]}
svr_model, svr_abs_preds = evaluate_model("Support Vector Regression (SVR)", SVR(), svr_params)

print("\nTuning XGBoost...")
xgb_params = {"n_estimators": [50, 100, 200], "learning_rate": [0.01, 0.05, 0.1], "max_depth": [3, 5, 7]}
_, _ = evaluate_model("XGBoost", xgb.XGBRegressor(random_state=42), xgb_params)

print("\nTuning Random Forest...")
rf_params = {"n_estimators": [50, 100, 200], "max_depth": [3, 5, 7, None]}
_, _ = evaluate_model("Random Forest", RandomForestRegressor(random_state=42), rf_params)
```

## 4. Statistical Significance
We run the non-parametric Wilcoxon Signed-Rank test to prove the SVR model's superiority is statistically significant.


```python
svr_errors = np.abs(actual_fcf - svr_abs_preds)
w_stat, p_w = wilcoxon(baseline_errors, svr_errors)
print(f"Wilcoxon signed-rank test p-value: {p_w:.5f}")
if p_w < 0.05:
    print("Conclusion: SVR is statistically significantly better than traditional forecasting! (p < 0.05)")
```

## 5. Explainable AI: Permutation Importance
SVR relies on a "black-box" non-linear Kernel Trick. To satisfy financial transparency requirements, we pry it open using Permutation Importance.


```python
result = permutation_importance(svr_model, X_train_scaled, y_train, n_repeats=10, random_state=42, n_jobs=-1)
importances = result.importances_mean
indices = np.argsort(importances)

plt.figure(figsize=(10, 6))
plt.barh(range(len(indices)), importances[indices], color="#2ca02c", align="center")
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.title("SVR Permutation Importance\n(What drives Hydropower Cash Flows?)", fontsize=14)
plt.xlabel("Mean Decrease in Accuracy")
plt.show()
```

## 6. Empirical True DCF Valuation (RQ3)
We utilize a 5-year projection mathematically derived from the SVR model and a 10% WACC. 
Because hydropower plants revert to the government in Nepal after 30 years, we bound the Terminal Value using a **Finite-Horizon approach** tied directly to each firm's `License_Term_Years`.

We then subtract Total Debt to calculate Equity Value, and compare it against the actual Stock Market Capitalization on the NEPSE!


```python
WACC = 0.10
PROJECTION_YEARS = 5

results_dcf = []
for _, row in test_df.iterrows():
    ticker = row["Ticker"]
    base_fcf = row["OperatingCashFlow_Proxy"]
    license_remaining = row["License_Term_Years"]
    actual_market_cap = row["Market_Cap"]
    total_debt = row["Total_Debt"]
    
    # ML Forecasting (Iterative)
    ml_forecasts = []
    current_fcf = base_fcf
    x_scaled = scaler.transform(pd.DataFrame([row[features]]))
    pred_growth = svr_model.predict(x_scaled)[0]
    
    for _ in range(PROJECTION_YEARS):
        next_fcf = current_fcf + (np.abs(current_fcf) + 1) * pred_growth
        ml_forecasts.append(next_fcf)
        current_fcf = next_fcf
        
    # Traditional Forecasting (0% Growth)
    trad_forecasts = [base_fcf for _ in range(PROJECTION_YEARS)]
    
    # Present Value of 5-year Cash Flows
    ml_pv = sum(fcf / ((1 + WACC) ** (t + 1)) for t, fcf in enumerate(ml_forecasts))
    trad_pv = sum(fcf / ((1 + WACC) ** (t + 1)) for t, fcf in enumerate(trad_forecasts))
    
    # Finite-Horizon Terminal Value
    remaining_years = max(0, int(license_remaining) - PROJECTION_YEARS)
    ml_tv_pv = 0
    trad_tv_pv = 0
    
    if remaining_years > 0:
        ml_tv = sum(ml_forecasts[-1] / ((1 + WACC) ** t) for t in range(1, remaining_years + 1))
        ml_tv_pv = ml_tv / ((1 + WACC) ** PROJECTION_YEARS)
        
        trad_tv = sum(trad_forecasts[-1] / ((1 + WACC) ** t) for t in range(1, remaining_years + 1))
        trad_tv_pv = trad_tv / ((1 + WACC) ** PROJECTION_YEARS)
        
    # Enterprise Value
    ml_ev = ml_pv + ml_tv_pv
    trad_ev = trad_pv + trad_tv_pv
    
    # Equity Value = Enterprise Value - Total Debt
    ml_eq = max(0, ml_ev - total_debt)
    trad_eq = max(0, trad_ev - total_debt)
    
    ml_error = np.abs(actual_market_cap - ml_eq)
    trad_error = np.abs(actual_market_cap - trad_eq)
    
    results_dcf.append({
        "Ticker": ticker, "SVR_Eq": ml_eq, "Trad_Eq": trad_eq, 
        "Actual_Cap": actual_market_cap, "SVR_Error": ml_error, "Trad_Error": trad_error
    })

res_df = pd.DataFrame(results_dcf)
avg_svr_err = res_df["SVR_Error"].mean()
avg_trad_err = res_df["Trad_Error"].mean()
improvement = ((avg_trad_err - avg_svr_err) / avg_trad_err) * 100

print(f"Average Traditional Valuation Error: Rs {avg_trad_err:,.0f}")
print(f"Average SVR Valuation Error:         Rs {avg_svr_err:,.0f}")
print(f"\nSVR beat the traditional stock market baseline by {improvement:.2f}%!")
```
